```python
#aggregating
#tolist-convert array to list
#unique-select unique value
code=data['CODE'].unique().tolist()
country=data['COUNTRY'].unique().tolist()
hdi=[]
tc=[]
td=[]
sti=[]
population=[]
gdp=[]

for x in country:
  hdi.append(data.loc[data['COUNTRY'] == x,'HDI'].sum()/294)
  tc.append(data.loc[data['COUNTRY'] == x,'TC'].sum())
  td.append(data.loc[data['COUNTRY'] == x,'TD'].sum())
  sti.append(data.loc[data['COUNTRY'] == x,'STI'].sum()/294)
  population.append(data.loc[data['COUNTRY'] == x,'POP'].sum())
  gdp.append(data.loc[data['COUNTRY'] == x,'GDPCAP'].sum())



aggregated_data = pd.DataFrame(list(zip(code, country,hdi,tc,td,sti,population,gdp))
                ,columns =['Country code', 'Country','HDI','Total Cases', 'Total Deaths', 'STI', 'Population','GDP'])
```

```python
aggregated_data
```

```python
# Sorting by column "x" -df.sort_values(by=['x'], ascending=False)
#df = aggregated_data

aggregated_data = aggregated_data.sort_values(by = ['Total Cases'],ascending = False)
```

```python
top_10_cases = aggregated_data.head(10)
```

```python
top_10_cases
```

```python
import plotly.express as px
fig = px.bar(top_10_cases, x='Country code', y='Total Cases',color='Country code', text_auto=True,
             title='Countries with highest covid cases')
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Bar(
    x=top_10_cases['Country'],
    y=top_10_cases['Total Cases'],
    name='Total Cases',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=top_10_cases['Country'],
    y=top_10_cases['Total Deaths'],
    name='Total Deaths',
    marker_color='lightsalmon'
))

fig.add_trace(go.Bar(
    x=top_10_cases['Country'],
    y=top_10_cases['Population'],
    name='Population',
    marker_color='black'
))

# Here we modify the tickangle of the xaxis, resulting in rotated labels.
fig.update_layout(barmode='group', xaxis_tickangle=-45, title = 'Total Cases vs Total Deaths')
fig.show()
```
